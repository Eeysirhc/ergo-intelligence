**Objective:** visualize price of $ERGOPAD by pulling statistics from ErgoDEX indexer  

## Load modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sqlalchemy import create_engine
```

## Retrieve on-chain data

```python
# default config
POSTGRES_ADDRESS = 'localhost' 
POSTGRES_PORT = '5432'
POSTGRES_USERNAME = 'ergodex' 
POSTGRES_PASSWORD = 'ergodex'
POSTGRES_DBNAME = 'ergodex'

postgres_str = ('postgresql://{username}:{password}@{ipaddress}:{port}/{dbname}'.format(username=POSTGRES_USERNAME,password=POSTGRES_PASSWORD,ipaddress=POSTGRES_ADDRESS,port=POSTGRES_PORT,dbname=POSTGRES_DBNAME))

conn = create_engine(postgres_str)

qry = open('../../sql/ergopad-ergodex_price.sql', 'r')
ergopad = pd.read_sql_query(qry.read(), conn)
qry.close()
```

```python
ergopad['100 Block Avg'] = ergopad['price'].rolling(window=100).mean()
ergopad.rename(columns={'price':'Price'}, inplace=True)
ergopad = ergopad[['global_index', 'Price', '100 Block Avg']] 
ergopad = ergopad.melt(id_vars='global_index')
ergopad.rename(columns={'variable':'segment'}, inplace=True)
```

```python
sns.set(style='darkgrid')

plt.clf()
plt.rcParams["figure.dpi"] = 100
fig = plt.figure(figsize=(20,15))

sns.lineplot(x='global_index', y='value', hue='segment', marker='o', markersize=10, data=ergopad, palette='Set2').invert_yaxis()
plt.ylim(125, 0)
plt.suptitle("ErgoPad: $ERGOPAD price over Ergo block time", x=0.12, y=0.91, fontsize=15, horizontalalignment='left')
plt.xlabel("Global Index")
plt.ylabel("ERGOPAD / ERG")

plt.show()
```
